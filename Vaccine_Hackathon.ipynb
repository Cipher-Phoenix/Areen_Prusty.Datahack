```python
y_train= pd.read_csv("training_set_labels.csv")
y_train=y_train.drop(["respondent_id"], axis=1)
y_train
```

```python
import xgboost as xgb
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import train_test_split

# Assuming you have X_train and y_train already loaded

# Split X_train and y_train into training and validation sets
X_train_split, X_val_split, y_train_split, y_val_split = train_test_split(X_train, y_train, test_size=0.2, random_state=42)

# Define parameters for XGBoost
params = {
    'objective': 'binary:logistic',   # Use logistic regression for binary classification
    'eval_metric': 'auc',             # Evaluation metric: ROC AUC
    'eta': 0.1,                       # Learning rate
    'max_depth': 5,                   # Maximum depth of each tree
    'subsample': 0.8,                 # Subsample ratio of the training instances
    'colsample_bytree': 0.8,          # Subsample ratio of columns when constructing each tree
    'seed': 42                        # Random seed for reproducibility
}

# Initialize XGBoost models
bst_first = xgb.XGBClassifier(**params)
bst_second = xgb.XGBClassifier(**params)

# Train the model for the first output feature
bst_first.fit(X_train_split, y_train_split['xyz_vaccine'])

# Train the model for the second output feature
bst_second.fit(X_train_split, y_train_split['seasonal_vaccine'])

# Predict probabilities on validation set
y_pred_proba_first = bst_first.predict_proba(X_val_split)[:, 1]
y_pred_proba_second = bst_second.predict_proba(X_val_split)[:, 1]

# Calculate ROC AUC for each output feature on validation set
auc_first = roc_auc_score(y_val_split['xyz_vaccine'], y_pred_proba_first)
auc_second = roc_auc_score(y_val_split['seasonal_vaccine'], y_pred_proba_second)

# Print ROC AUC scores
print("ROC AUC for first output feature (xyz_vaccine):", auc_first)
print("ROC AUC for second output feature (seasonal_vaccine):", auc_second)
```

```python
import xgboost as xgb
from sklearn.metrics import roc_auc_score


dtrain_first = xgb.DMatrix(X_train, label=y_train['xyz_vaccine']) 
dtrain_second = xgb.DMatrix(X_train, label=y_train['seasonal_vaccine'])  

# Define parameters for XGBoost
params = {
    'objective': 'binary:logistic',   # Use logistic regression for binary classification
    'eval_metric': 'auc',             # Evaluation metric: ROC AUC
    'eta': 0.1,                       # Learning rate
    'max_depth': 5,                   # Maximum depth of each tree
    'subsample': 0.8,                 # Subsample ratio of the training instances
    'colsample_bytree': 0.8,          # Subsample ratio of columns when constructing each tree
    'seed': 42                        # Random seed for reproducibility
}

# Train the model for the first output feature
num_round = 100  # Number of boosting rounds (iterations)
bst_first = xgb.train(params, dtrain_first, num_round)

# Train the model for the second output feature
bst_second = xgb.train(params, dtrain_second, num_round)

# Make predictions on the training set for both output features
y_pred_proba_first = bst_first.predict(dtrain_first)
y_pred_proba_second = bst_second.predict(dtrain_second)

# Example: Assuming y_train is the true labels for X_train
# Replace y_train with your actual true labels if available
y_true_first = y_train['xyz_vaccine']
y_true_second = y_train['seasonal_vaccine']

# Calculate ROC AUC for each output feature on the training set
auc_first = roc_auc_score(y_true_first, y_pred_proba_first)
auc_second = roc_auc_score(y_true_second, y_pred_proba_second)

# Print or use the ROC AUC scores as needed
print("ROC AUC for first output feature on training set:", auc_first)
print("ROC AUC for second output feature on training set:", auc_second)




# Convert X_test into DMatrix format for XGBoost (for testing)
dtest = xgb.DMatrix(X_test)

# Make predictions on X_test for both output features
y_pred_proba_first = bst_first.predict(dtest)
y_pred_proba_second = bst_second.predict(dtest)

# Example: Print or use the predictions as needed
print("Predictions for first output feature on X_test:")
print(y_pred_proba_first)

print("\nPredictions for second output feature on X_test:")
print(y_pred_proba_second)
```
